## Dataset Overview

The dataset contains the following columns:

* **ID**                                                                                                             : Unique identifier for each record.
* **Gender**                                                                                                         : Gender of the applicant (binary: 0 or 1).
* **Has a car**                                                                                                      : Indicates whether the applicant owns a car (binary: 0 or 1).
* **Has a property**                                                                                                 : Indicates whether the applicant owns a property (binary: 0 or 1).
* **Children count**                                                                                                 : Number of children the applicant has.
* **Income**                                                                                                         : Annual Income of the applicant.
* **Employment status**                                                                                              : Employment status of the applicant.
* **Education level**                                                                                                : Highest education level attained by the applicant.
* **Marital status**                                                                                                 : Marital status of the applicant.
* **Dwelling**                                                                                                       : Type of dwelling the applicant resides in.
* **Age**                                                                                                            : Age of the applicant in days. Counted backwards from the current day.
* **Employment length**                                                                                              : How long the customer has been employed (days), counted backwards from the current day. Positive numbers indicate unemployment.
* **Has a mobile phone**                                                                                             : Indicates whether the applicant has a mobile phone (binary: 0 or 1).
* **Has a work phone**                                                                                               : Indicates whether the applicant has a work phone (binary: 0 or 1).
* **Has a phone**                                                                                                    : Indicates whether the applicant has any phone (binary: 0 or 1).
* **Has an email**                                                                                                   : Indicates whether the applicant has an email (binary: 0 or 1).
* **Job title**                                                                                                      : Title or position of the applicant's job.
* **Family member count**                                                                                           : Number of family members.
* **Account age**                                                                                                     : Age of the applicant's account in years. Counted backwards from the current day.
* **Is high risk (Target Variable)**: Whether the applicant is considered high risk (0 for no, 1 for yes).

#### Age

The age column need to be preprocessed and it's values should be changed before doing any analysis on it.

```python
# Todays date
today = date.today()

# Creating an function to modify the Age Column values.
def Age(age):    
    daysBack = int(age)    
    dob = (today + timedelta(days= daysBack))
    return today.year - dob.year

# Modifying Age column values to how old they are in years.
trainDF["Age"] = trainDF["Age"].apply(Age)
```

#### Employment length

```python
# Todays date
today = date.today()

# Creating an function to modify the Age Column values.
def EmploymentLength(empLen):
    if empLen > 0:
        return -1
    
    daysBack = int(empLen)    
    joiningDate = (today + timedelta(days= daysBack))
    return today.year - joiningDate.year

# Modifying Age column values to how old they are in years.
trainDF["Employment length"] = trainDF["Employment length"].apply(EmploymentLength)
trainDF.rename(columns= {"Employment length": "Employment length in years"}, inplace= True)
```

#### Account Age

```python
# Converting Negative Account age to +ve.
trainDF["Account age"] = trainDF["Account age"].apply(lambda x: int(-1 * x))
```

##### Education level

```python
trainDF["Education level"].value_counts()
```

```python
# Merging Last 3 Categories as other

def renameEducationLevel(eduLevel):
    if eduLevel == "Incomplete higher":
        return "Other"
    elif eduLevel == "Lower secondary":
        return "Other"
    elif eduLevel == "Academic degree":
        return "Other"
    else:
        return eduLevel

trainDF["Education level"] = trainDF["Education level"].apply(renameEducationLevel)
```

```python
trainDF["Education level"].value_counts()
```

##### Martial Status

```python
trainDF["Marital status"].value_counts()
```

```python
# Merging Last 2 Categories as other

def renameMaritalStatus(status):
    if status == "Separated":
        return "Other"
    elif status == "Widow":
        return "Other"
    else:
        return status

trainDF["Marital status"] = trainDF["Marital status"].apply(renameMaritalStatus)
```

```python
trainDF["Marital status"].value_counts()
```

```python
trainDF["Dwelling"].value_counts()
```

```python
trainDF["Dwelling"] = trainDF["Dwelling"].apply(lambda x: x if x== "House / apartment" else "Other")
```

##### Job Title

```python
trainDF["Job title"].value_counts()
```

```python
# Updating all the values of the job Titles column from nan to Unemployed.
trainDF.loc[trainDF["Employment length in years"] == -1, "Job title"] = "Unemployed"

# Setting other Job Title Values to Other
otherJobTitle = trainDF["Job title"].value_counts().index[8:]

def renameJobTitle(jobTitle):
    if jobTitle in otherJobTitle:
        return "Other"

    elif jobTitle is np.NaN:
        return "Other"
        
    else:
        return jobTitle

# Updating all the Job Title's values
trainDF["Job title"] = trainDF["Job title"].apply(renameJobTitle)
```

## Creating Pipeline

```python
# One Hot Encoding Transformers

oheTrf = ColumnTransformer([
    ("ohe_trf_cat_cols", OneHotEncoder(sparse_output= False, handle_unknown= "ignore", drop="first"), [0, 1, 2, 5, 6, 7, 8, 14])
], remainder= "passthrough")
```

```python
# Standard Scaler Transformer

stdScalerTrf = ColumnTransformer([
    ("std_scaler_trf", StandardScaler(), [4])
], remainder= "passthrough")
```

```python
# Creating a Model Transformer.
modelTrf = DecisionTreeClassifier(max_depth= 120, 
                            criterion= "entropy",
                            min_samples_split= 2, 
                            min_samples_leaf= 2)
```

```python
# Creating a pipeline and assigning the transforms in the pipeline
pipe = Pipeline([
    ("ohe_trf_cat_cols", oheTrf),
    ("std_scaler_trf", stdScalerTrf),
    ("modelTrf", modelTrf)
])
```

```python
# Training the model
pipe.fit(x_train, y_train)

# Predicting the values
y_pred = pipe.predict(x_test)
```

```python
# Printing The Metrices
print("Accuracy:- ", accuracy_score(y_test, y_pred))
print("Precision:- ", precision_score(y_test, y_pred))
print("Recall:- ", recall_score(y_test, y_pred))
print("F1 Score:- ", f1_score(y_test, y_pred))
print("Confusion Matrix:- ", confusion_matrix(y_test, y_pred))
print("roc_auc_score:- ", roc_auc_score(y_test, y_pred))
```
