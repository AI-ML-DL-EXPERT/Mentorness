```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming 'date' is the datetime column for the time series
df.set_index('Order Date', inplace=True)

# Extract the sales column
sales = df['Sales']

# Split data into training and testing sets (e.g., 80% for training, 20% for testing)
train_size = int(len(sales) * 0.8)
train, test = sales[:train_size], sales[train_size:]


# 1. Visualize the time series data
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Testing Data')
plt.title('Furniture Sales Time Series')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()


# 2. Decompose time series to explore trends and seasonality
decomposition = seasonal_decompose(train, model='additive', period=12)  # Adjust period if needed
decomposition.plot()
plt.tight_layout()
plt.show()


# 3. Simple Moving Average (SMA) for short-term forecasting
window_size = 3  # Adjust window size as needed
sma = train.rolling(window=window_size).mean()  # Calculate SMA

# Plot actual vs SMA forecast
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Actual Sales')
plt.plot(sma.index, sma, label=f'SMA (Window: {window_size})')
plt.title('Furniture Sales - SMA Forecast')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# 4. Visualization with Seaborn (example)
sns.lineplot(data=df, x='Order Date', y='Sales')  # Adjust based on your DataFrame structure
plt.title('Furniture Sales Time Series (Seaborn)')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_furniture_sales_timeseries(df, sales_column="Sales"):
  """
  Plots the furniture sales time series data with clear formatting and informative elements.

  This function creates a line chart with clear axis labels, gridlines, and potentially color highlights
  to effectively visualize the furniture sales trends over time.

  Args:
      df (pandas.DataFrame): The DataFrame containing the time series data.
          It's assumed to have a datetime index named 'Order Date' and a column named 'Sales' for sales data.
      sales_column (str, optional): The name of the column containing sales data. Defaults to "Sales".
  """

  # df.set_index('Order Date', inplace=True)  # Assuming 'Order Date' is the datetime column
  sales = df[sales_column]

  # Create the time series plot with informative elements
  plt.figure(figsize=(12, 6))
  plt.plot(sales.index, sales, color='blue', label='Furniture Sales')
  plt.title('Furniture Sales Time Series')
  plt.xlabel('Order Date')
  plt.ylabel('Sales')
  plt.legend()
  plt.grid(True)
  plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
  plt.tight_layout()
  plt.show()


# Example Usage (assuming your DataFrame is named 'df')
plot_furniture_sales_timeseries(df.copy())  # Pass a copy to avoid modifying the original DataFrame
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def plot_furniture_sales_heatmap(df, sales_column="Sales", resample_freq="ME"):
  """
  Visualizes furniture sales data as a heatmap to explore seasonal patterns.

  This function resamples the data by month and creates a heatmap
  where the color intensity represents the sales volume for each month across different years.

  Args:
      df (pandas.DataFrame): The DataFrame containing the time series data.
          It's assumed to have a datetime index named 'Order Date' and a column named 'Sales' for sales data.
      sales_column (str, optional): The name of the column containing sales data. Defaults to "Sales".
      resample_freq (str, optional): The frequency for resampling the data (e.g., 'ME' for monthly endpoints). Defaults to "ME".
  """

  # df.set_index('Order Date', inplace=True)  # Assuming 'Order Date' is the datetime column
  sales = df[sales_column]

  # Resample data by month (adjust as needed)
  monthly_sales = sales.resample(resample_freq).sum()  # Calculate monthly sum

  # Create the heatmap (using monthly_sales directly)
  plt.figure(figsize=(10, 6))
  sns.heatmap(monthly_sales.to_frame().transpose(), cmap="coolwarm", annot=True, fmt=".0f")  # Adjust colormap and formatting
  plt.title('Furniture Sales Heatmap (Monthly)')
  plt.xlabel('Year')
  plt.ylabel('Month')
  plt.xticks(rotation=90)  # Ensure year labels are horizontal
  plt.tight_layout()
  plt.show()


# Example Usage (assuming your DataFrame is named 'df')
plot_furniture_sales_heatmap(df.copy())  # Pass a copy to avoid modifying the original DataFrame
```
